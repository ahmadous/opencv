```python
#Exercice 1
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parametre du signal
f0=400
duree=2
Fe=100
```

```python
#creation du vecteur temps
t=np.linspace(0,duree,int(Fe * duree))
```

```python
#creation du signal sinusoidal
signal=np.sin(2 * np.pi * f0* t)
```

```python
#representation du signal
plt.plot(t,signal)
plt.title("signal sinusoidal")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
